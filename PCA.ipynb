```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Create a PCA model with 2 components
pca = PCA(n_components=2) # PCA reduces the 4-dimensional data (since Iris has 4 features) into 2 dimensions.
```

```python
# Fit and transform the data to the first 2 principal components
X_r = pca.fit_transform(X) # Learns the directions of maximum variance (principal components) and projects the original 4-dimensional data into 2 dimensions.
```

```python
# Plot the reduced data
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
  plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
  plt.legend(loc='best', shadow=False, scatterpoints=1)
  plt.title('PCA of IRIS dataset')
  plt.show()
```
