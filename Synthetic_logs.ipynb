CLUSTERING AND TOKENIZATION

```python
from sklearn.cluster import DBSCAN
from sentence_transformers import SentenceTransformer
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings=model.encode(df.log_message.tolist())
```

```python
df.head(1)
```

```python
embeddings[0:5]
```

```python
clustering=DBSCAN(eps=0.2, min_samples=1, metric="cosine").fit(embeddings)
df['cluster']=clustering.labels_
df['cluster'].unique()
```

```python
df.sample(10)
```

```python
df[df.cluster==0].sample(5)
```

GROUP BY CLUSTERS

```python
clusters=df.groupby('cluster')['log_message'].apply(list)
sorted_clusters=clusters.sort_values(key=lambda x: x.map(len), ascending=False)
```

```python
print("Clustered patterns")

for cluster_id,messages in sorted_clusters.items():
  if len(messages)>10:
    print(f"Cluster {cluster_id}:")
    for message in messages[:5]:
      print(f"{message}")
```

```python
for cluster_id, messages in sorted_clusters.items():
    if len(messages) > 10:
        print(f"Cluster {cluster_id}:")

        # Optionally, you can sort messages by some attribute (e.g., length of message)
        sorted_messages = sorted(messages)  # or use a custom sorting criteria

        # Display the first 5 messages of the sorted messages
        for message in sorted_messages[:5]:
            print(f"{message}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
embeddings_filtered=model.encode(df.log_message.tolist())
```

```python
x=embeddings_filtered
y=df['target_label'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
clf=LogisticRegression(max_iter=1000)
clf.fit(x_train,y_train)
```

```python
x_test.shape
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
